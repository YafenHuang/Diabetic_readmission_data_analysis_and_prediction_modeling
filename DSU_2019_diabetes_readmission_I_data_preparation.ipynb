## I. Diabetes Readmission data set initial data cleansing

### Steps
#### 1. data import
#### 2. Remove duplicates
#### 3. Replace missing value '?' with NaN
#### 4. Drop features with too many missing values
#### 5. Add full description of some features

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
import numpy as np
from scipy import stats
```

###  1.Import oringinal data

```python
diabetic_data_initial=pd.read_csv(r'D:\Yafen\study\UHG DSU\my project\dataset_diabetes\diabetic_data.csv')
```

```python
print('diabetic_data_initial shape:',diabetic_data_initial.shape)
diabetic_data_initial.head()
```

### 2. Data Cleansing
#### Overview of data in original dataset 

```python
diabetic_data_initial.describe(include='all')
```

#### Replace missing value '?' with NaN

```python
diabetic_data_initial.replace('?',np.nan,inplace=True)
```

```python
diabetic_data_initial.head()
```

```python
diabetic_data_initial_missing_value_summary=pd.DataFrame(diabetic_data_initial.isnull().sum(axis=0))
diabetic_data_initial_missing_value_summary.reset_index(inplace=True)
diabetic_data_initial_missing_value_summary
diabetic_data_initial_missing_value_summary.columns=['feature','missing_value_count']
diabetic_data_initial_missing_value_summary['missing_value_rate']=round(diabetic_data_initial_missing_value_summary['missing_value_count']/101766*100,2)
diabetic_data_initial_missing_value_summary#.sort_values('missing_value_rate',ascending=False)                                                                      
```

### Drop features that contain too many missing values: weight

```python
diabetic_data_clean=diabetic_data_initial.drop(columns=['weight'])
```

```python
diabetic_data_clean.shape
```

### Check patients with multiple encounters and only keep one encounter per one patient

```python
diabetic_data_clean.drop_duplicates('encounter_id',keep='first',inplace=True)
diabetic_data_clean.shape
```

```python
101766-71518
```

```python
diabetic_data_clean.drop_duplicates('patient_nbr',keep='first',inplace=True)
diabetic_data_clean.shape
```

### Remove encounters that resulted in death 

```python
discharge_disposition_id=pd.read_csv(r'D:\Yafen\study\UHG DSU\my project\dataset_diabetes\discharge_disposition_id.csv')
discharge_disposition_id
```

```python
diabetic_data_clean['discharge_disposition']=pd.merge(discharge_disposition_id,diabetic_data_clean,left_on='discharge_disposition_id',right_on='discharge_disposition_id')['description_revised']
diabetic_data_clean=diabetic_data_clean[diabetic_data_clean['discharge_disposition']!='hospice']
```

```python
print('Diabetic data after cleansing',diabetic_data_clean.shape)
```

### 3. Adding description 


#### Map admission_type

```python
admission_type_id=pd.read_csv(r'D:\Yafen\study\UHG DSU\my project\dataset_diabetes\admission_type_id.csv')
admission_type_id.head()
```

```python
diabetic_data_clean['admission_type']=pd.merge(admission_type_id,diabetic_data_clean,left_on='admission_type_id',right_on='admission_type_id')['description_revised']
```

```python
diabetic_data_clean.groupby('admission_type').size()
```

```python
admission_source_id=pd.read_csv(r'D:\Yafen\study\UHG DSU\my project\dataset_diabetes\admission_source_id.csv')
admission_source_id.head()
```

```python
diabetic_data_clean['admission_source']=pd.merge(admission_source_id,diabetic_data_clean,left_on='admission_source_id',right_on='admission_source_id')['description']
```

```python
diabetic_data_clean.groupby(['admission_type','admission_source']).size()
```

```python
diabetic_data_clean.drop(columns=['admission_type_id','discharge_disposition_id','admission_source_id','encounter_id','patient_nbr','payer_code','medical_specialty'],inplace=True)
```

### Export data 

```python
diabetic_data_clean.to_csv(r'D:\Yafen\study\UHG DSU\my project\dataset_diabetes\diabetic_data_clean.csv')
```
